```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
path = '../data/hrdata.csv'
```

```python
df = pd.read_csv(path, dtype={'id': str})
```

```python
df.head()
```

```python
# convert to datetime 
df['hired_date'] = pd.to_datetime(df['hired_date'])
```

# Task 1

```python
df['end_of_month'] = pd.to_datetime(df['hired_date'],'%Y-%m' ) + MonthEnd(1)
```

```python
df['end_of_month'] = df['end_of_month'].dt.date
```

```python
df
```

# Task 2

```python
df['current_date'] = dt.now()
```

```python
df['current_date'] = df['current_date'].dt.date
```

```python
df.loc[:, df.columns!='end_of_month']
```

```python
df['service_with_us'] = pd.to_datetime(df['current_date']) - df['hired_date']
```

```python
df['service_with_us(yr)'] = df['service_with_us']/np.timedelta64(1,'Y')
```

```python
df.loc[:, ~df.columns.isin(['end_of_month', 'current_date', 'service_with_us'])]
```

```python
df.loc[:, ~df.columns.isin(['end_of_month', 'current_date', 'service_with_us'])]
```

# Task 3

```python
df.loc[df['service_with_us(yr)'] > 3, 'Bonus'] = df['salary'] * 0.4
```

```python
df.loc[:, ~df.columns.isin(['end_of_month', 'current_date', 'service_with_us'])]
```

```python
df.to_csv("../data/hrdataupdate.csv")
```

```python
df.groupby('Sex').count().reset_index()
```

```python
counts = df.groupby('Sex').count().reset_index()[['Sex', 'id']]
```

```python
counts
```

```python
total = counts['id'].sum(axis=0)
```

```python
male_percentage = counts.loc[counts['Sex'] == 'M']['id']/total * 100
```

```python
female_percentage = counts.loc[counts['Sex'] == 'F']['id']/total * 100
```

```python
print(f"Male percentage : {male_percentage.values[0]} \n Female percentage : {female_percentage.values[0]}")
```
